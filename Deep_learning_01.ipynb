```python
from transformers import pipeline
```

```python
#text classification
```

```python
pipe=pipeline(task="sentiment-analysis",model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
text = "one day I will be richest man on the world"
```

```python
pipe(text)
```

```python
# Zero-Short Text Classification
```

```python
new_pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
text="i love to see the world and also like cooking and camping"
candidate_labels = ['travel', 'cooking', 'dancing', 'adventure']
```

```python
new_pipe(text,candidate_labels)
```

```python
##Deep Learning
```

```python
import numpy as np
```

```python
from tensorflow.keras import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
xs = np.array([-1, 0, 1, 2, 3, 4])
ys = np.array([-3, -1, 1, 2.8, 5.1, 6.9])
```

```python
model=Sequential()
```

```python
model.add(Dense(units=1, input_shape=[1]))
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(xs, ys, epochs=500)
```

```python
model.predict([10])
```
